```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
import tensorflow.keras.preprocessing.text as tfkpt
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
scotch_df = pd.read_csv('scotch_review.csv')
```

```python
scotch = scotch_df.description
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(scotch)
```

```python
scotch = tokenizer.texts_to_sequences(scotch)
scotch
```

```python
scotch = tokenizer.sequences_to_matrix(scotch)
scotch
```

```python
the_review = scotch_df[['description', 'review.point']].sort_values(by='review.point')
the_review
```

```python
def create_sentiment(int):
    if int > 89 :
        return 2
    if int < 89 and int > 79:
        return 1
    if int < 79:
        return 0
```

```python
the_review['sentiment'] = the_review['review.point'].apply(create_sentiment)
the_review.sample(5)
```

```python
def replace_commas(str):
    return str.replace(',',';')
```

```python
the_review['description'] = the_review['description'].apply(replace_commas)
the_review.sample(5)
```
